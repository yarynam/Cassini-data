```python
# global vars to calculate moon's orbit
step = 1000 #1000 posiitions of the planet during the orbital period 
end_date  = datetime.strptime('2017-09-15T10:32:00', "%Y-%m-%dT%H:%M:%S")
```

```python
moons_data = []

for i, planet in enumerate(planets):
    
    # get initial position 
    init_position, init_lightTime = spice.spkpos(planet["name"], init_date, 'J2000', 'NONE', 'SATURN')
    
    if planet["name"] == "SATURN":
        planet["radius"] = planet["radius"]/1000
        planet["pos_x"] = 0
        planet["pos_y"] = 0
        planet["pos_z"] = 0
        planet["class"] = "moon"
    else:
        #get orbit coordinates 
        orbital_period = planet["orbital_period"] + (planet["orbital_period"]/10)
        start_date = end_date - timedelta(hours = orbital_period) #end date of orbital period
        utc = [start_date.strftime("%Y-%m-%dT%H:%M:%S"), end_date.strftime("%Y-%m-%dT%H:%M:%S")]

        etOne = spice.str2et(utc[0])
        etTwo = spice.str2et(utc[1])

        times = [x*(etTwo-etOne)/step + etOne for x in range(step)]
        times_utc = map(lambda t : spice.timout(t, "YYYY-MM-DD HR:MN:SC", 20), times)


        positions, lightTimes = spice.spkpos(planet["name"], times, 'J2000', 'NONE', 'SATURN')
        orbit_positions = []
        
        # position on orbit
        for k, old_dict in enumerate(positions):
            new_dict = {"pos_x":old_dict[1]/1000, 
                        "pos_y":old_dict[2]/1000,
                        "pos_z":old_dict[0]/1000,
                        "time": times_utc[k]}
            orbit_positions.append(new_dict)

        # initial position and radius in 1000
        planet["pos_x"] = init_position[1]/1000 
        planet["pos_y"] = init_position[2]/1000
        planet["pos_z"] = init_position[0]/1000
        planet["radius"] = planet["radius"]/1000
        planet["orbit"] = orbit_positions
        planet["class"] = "moon"

    
        
        
print(planets[1])
    
#     #get orbit coordinates 
#     end_d = start_d + timedelta(hours= moons_op[i]) #end date of orbital period
#     utc = [start_d.strftime("%Y-%m-%dT%H:%M:%S"), end_d.strftime("%Y-%m-%dT%H:%M:%S")]
    
#     etOne = spice.str2et(utc[0])
#     etTwo = spice.str2et(utc[1])
    
#     times = [x*(etTwo-etOne)/step + etOne for x in range(step)]
#     times_utc = map(lambda t : spice.timout(t, "YYYY-MM-DD HR:MN:SC", 20), times)
    
    
#     positions, lightTimes = spice.spkpos(moon, times, 'J2000', 'NONE', 'SATURN')
#     orbit_positions = []

#     for k, old_dict in enumerate(positions):
#         new_dict = {"pos_x":old_dict[1]/1000, 
#                     "pos_y":old_dict[2]/1000,
#                     "pos_z":old_dict[0]/1000,
#                     "time": times_utc[k]}
#         orbit_positions.append(new_dict)
    
#     moon_data = {"init_pos_x":init_position[1]/1000, 
#                 "init_pos_y":init_position[2]/1000,
#                 "init_pos_z":init_position[0]/1000,
#                 "time": "2017-09-15 10:45:00" ,
#                 "name": moon,
#                 "radius": moons_r[i]/1000,
#                 "orbit": orbit_positions }
#     moons_data.append(moon_data)
    
# print moons_data[0]["orbit"]
```
