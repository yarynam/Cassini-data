```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import json #Import json lib to get nice ad valid json 
from datetime import datetime

import spiceypy as spice
spice.tkvrsn("TOOLKIT")
```

```python
spice.furnsh("./cassMetaK.txt")
```

```python
step = 20000
# we are going to get positions between these two dates
utc = ['2017-04-23T03:46:00', '2017-09-15T10:45:00']

# get et values one and two, we could vectorize str2et
etOne = spice.str2et(utc[0])
etTwo = spice.str2et(utc[1])

print("ET One: {}, ET Two: {}".format(etOne, etTwo))
```

```python
# get times
times = [x*(etTwo-etOne)/step + etOne for x in range(step)]

# get utc times
times_utc = map(lambda t : spice.timout(t, "YYYY-MM-DD HR:MN:SC", 20), times)
```

```python
#Run spkpos as a vectorized function
positions, lightTimes = spice.spkpos('Cassini', times, 'J2000', 'NONE', 'SATURN')


#Detect orbit number
def f(x):
    if datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-09-15 05:37:00", '%Y-%m-%d %H:%M:%S'): return 293
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-09-05 18:47:00", '%Y-%m-%d %H:%M:%S'): return 292
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-08-30 07:48:00", '%Y-%m-%d %H:%M:%S'): return 291
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-08-23 08:51:00", '%Y-%m-%d %H:%M:%S'): return 290
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-08-17 15:23:00", '%Y-%m-%d %H:%M:%S'): return 289
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-08-10 22:56:00", '%Y-%m-%d %H:%M:%S'): return 288
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-08-04 11:46:00", '%Y-%m-%d %H:%M:%S'): return 287
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-07-29 00:33:00", '%Y-%m-%d %H:%M:%S'): return 286
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-07-22 13:27:00", '%Y-%m-%d %H:%M:%S'): return 285
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-07-16 02:21:00", '%Y-%m-%d %H:%M:%S'): return 284
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-07-09 15:13:00", '%Y-%m-%d %H:%M:%S'): return 283
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-07-03 03:54:00", '%Y-%m-%d %H:%M:%S'): return 282
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-06-26 16:35:00", '%Y-%m-%d %H:%M:%S'): return 281
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-06-20 05:26:00", '%Y-%m-%d %H:%M:%S'): return 280
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-06-13 18:24:00", '%Y-%m-%d %H:%M:%S'): return 279
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-06-07 07:18:00", '%Y-%m-%d %H:%M:%S'): return 278
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-05-31 20:04:00", '%Y-%m-%d %H:%M:%S'): return 277
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-05-25 08:50:00", '%Y-%m-%d %H:%M:%S'): return 276
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-05-18 21:59:00", '%Y-%m-%d %H:%M:%S'): return 275
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-05-12 11:30:00", '%Y-%m-%d %H:%M:%S'): return 274
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-05-06 00:59:00", '%Y-%m-%d %H:%M:%S'): return 273
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') > datetime.strptime("2017-04-29 14:22:00", '%Y-%m-%d %H:%M:%S'): return 272
    elif datetime.strptime(x, '%Y-%m-%d %H:%M:%S') >= datetime.strptime("2017-04-23 03:46:00", '%Y-%m-%d %H:%M:%S'): return 271
    
    else: return 0

custom_position = []

for i, old_dict in enumerate(positions):
    new_dict = {"pos_x":old_dict[1]/1000, 
                "pos_y":old_dict[2]/1000,
                "pos_z":old_dict[0]/1000,
                "time": times_utc[i],
                "orbit": f(times_utc[i])}
    custom_position.append(new_dict)

# for old_list in positions:
#     new_list = map(lambda p : p*3/60268, old_list)
#     custom_position.append(new_list)
    
# Light times is a N vector of time
# print("Light Times: ")
# print(lightTimes[0])

print(custom_position[0])
print(custom_position[160])
print(custom_position[19890])
# with open("list.txt", "w") as f:
#     f.write("%s" % json.dumps(custom_position)) #write json

```
